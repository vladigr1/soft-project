# Lab4 - compare between the algorithems

# StepWise

```python
import algorithms


in_threshold = 0.3
out_threshold = 0.05
sderived = 'amount of sleep'
data = algorithms.load_data('30days.csv', sderived)

lbasics = list(data.columns.values)
lbasics.remove(sderived)

amount_of_days = 15

# print(data_15_days)

from algorithms import stepwise_regression
import evaluation

(fit_model, param_in_model, cur_pvalue) = stepwise_regression(data, sderived, lbasics, in_threshold, out_threshold)
print(f'param_in_model: {param_in_model}')
print(f'cur_pvalue: {cur_pvalue}')
distance = evaluation.distance_vec(data, sderived, param_in_model, fit_model)
print(f'accuracy: {evaluation.accuracy(distance)}')

x, y = evaluation.evaluation_graph(data, sderived, lbasics, amount_of_days, stepwise_regression )
import numpy as np
import matplotlib.pyplot as plt


N = len(x)
c = range(N)

fig, ax = plt.subplots()

scatter = ax.scatter(x, y, c=c)
plt.xlabel("p-value")
plt.ylabel("accuracy")

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                loc="upper right", title="Model number")
ax.add_artist(legend1)
plt.show()
```
